### How-to-count-distance-to-the-previous-zero

For each value, count the difference of the distance from the previous zero (or the start of the Series, whichever is closer) and if there are no previous zeros, print the position<br><br>
Consider a DataFrame df where there is an integer column {'X'[7, 2, 0, 3, 4, 2, 5, 0, 3, 4]}<br>
The values should therefore be [1, 2, 0, 1, 2, 3, 4, 0, 1, 2]. Make this a new column 'Y'.<br><br>
import pandas as pd<br><br>
df = pd.DataFrame({'X': [7, 2, 0, 3, 4, 2, 5, 0, 3, 4]})<br><br>

```python
df = pd.DataFrame({'X': [7, 2, 0, 3, 4, 2, 5, 0, 3, 4]}); 
print (type(df))

lst=[]
idx = 0
for index, row in df.iterrows():
    if row["X"] == 0:
        idx = 0
    else:
        idx+=1
    lst.append(idx)

df["Y"] = lst
df
```

### Create a DatetimeIndex that contains each business day of 2015 and use it to index a Series of random numbers.

```python
ind_x = pd.date_range("2015-01-01", "2015-12-31", freq='B')
```

```python
s = pd.Series(np.random.randint(1,100, len(ind_x)), index=ind_x);s
```

### Find the sum of the values in s for every Wednesday

```python
sum(s[s.index.dayofweek == 2])        # Day of week is part of pandas DateTimeIndex. 0 is Monday and 6 is Sunday
```

### Average For each calendar month

```python
s.groupby(s.index.month).sum()
```

### For each group of four consecutive calendar months in s, find the date on which the highest value occurred.

```python
s.groupby(pd.Grouper(freq='4MS')).max().dropna()
```
