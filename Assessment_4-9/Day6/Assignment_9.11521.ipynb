### We have the min and max temperatures in a city In India for each months of the year.

We would like to find a function to describe this and show it graphically, the dataset given below.<br>

Data<br>
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25<br>
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18<br>

```python
fig = plt.figure()
axes= fig.add_axes([0.1, 0.1, 0.8, 0.8])
                    
# Create the data sets
max = [39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]
min = [21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]
index = [i for i in range(0, len(max))]

# plot the graphs
axes.plot(index, max,'ro')
axes.plot(index, min,'bo')
axes.set_xlabel("Month")
axes.set_ylabel("Temperature")
```

Charts to plot:<br>
1. Create a pie chart presenting the male/female proportion<br>
2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender<br>

```python
url="https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
titanic = pd.read_csv(url)
titanic.head()
```

```python
label = ["Female", "Male"]
sizes = list(titanic.sex.value_counts().sort_index()); 
color = ['gold', 'yellowgreen']
# Plot
plt.pie(sizes, labels=label, colors=color, shadow=True, startangle=0, autopct='%1.1f%%')
plt.axis('equal')
```

### Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
plt.scatter(titanic[titanic["sex"] == "female"]["age"], titanic[titanic["sex"] == "female"]["fare"], color="red", label="Female")
plt.scatter(titanic[titanic["sex"] == "male"]["age"], titanic[titanic["sex"] == "male"]["fare"], color="blue", label="Male")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(loc=2)
```

```python
# Use seaborn scatter plot
import seaborn as sns
sns.set_style("whitegrid");
sns.FacetGrid(titanic, hue='sex', height=6)\
    .map(plt.scatter, "age", "fare")\
    .add_legend();
```
