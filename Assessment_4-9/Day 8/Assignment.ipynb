### Problem statement 2:

Read the following data set:<br>
https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data

```python
import requests
from io import StringIO

u1 = 'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
r1 = requests.get(u1, verify=False)
adult_data = pd.read_csv(StringIO(r1.text), header=None)
```

```python
adult_data.head()
```

```python
# Append column names to the table for ease of execution
adult_data.columns = user_columns.append(pd.Series(['NA','NA']), ignore_index=True)[0]
```

### 1. Create an sqlalchemy engine using a sample from the data set

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite://')
```

```python
adult_data.to_sql('user_data', con=engine)      # here user_data is the table name associated with the DataFrame
```

### 2. Write two basic update queries

```python
sql = 'update user_data set "native-country" = " USA" where "native-country" = " United-States"'
with engine.begin() as conn:
    conn.execute(sql)
    
engine.execute('select * from user_data limit 10').fetchall()
```

```python
sql = 'update user_data set NA = "less than 50K" where NA = " <=50K"'
with engine.begin() as conn:
    conn.execute(sql)
    
engine.execute('select * from user_data limit 10').fetchall()
```

### 3. Write two delete queries

```python
sql = 'delete from user_data where NA = "less than 50K"'
with engine.begin() as conn:
    conn.execute(sql)
    
engine.execute('select * from user_data where NA = "less than 50K"').fetchall()
```

```python
sql = 'delete from user_data where sex = " Male"'
with engine.begin() as conn:
    conn.execute(sql)
    
engine.execute('select * from user_data where sex = " Male"').fetchall()
```

### 4. Write two filter queries

```python
engine.execute('select * from user_data where sex = " Male"').fetchall()
```

```python
engine.execute('select * from user_data where race = " White" and sex = " Male"').fetchall()
```

### 5. Write two function queries

```python
engine.execute('select age, count(*) from user_data group by age').fetchall()
```
